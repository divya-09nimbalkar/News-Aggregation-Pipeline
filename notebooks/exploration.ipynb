```python
import matplotlib.pyplot as plt

def plot_sentiment_distribution(df):
    """
    Plots histogram of sentiment scores.
    """
    plt.figure(figsize=(6,4))
    df["sentiment"].plot(kind="hist", bins=10, color="skyblue", edgecolor="black")
    plt.title("Sentiment Distribution")
    plt.xlabel("Sentiment Score")
    plt.ylabel("Frequency")
    plt.show()

def plot_articles_by_source(df):
    """
    Plots bar chart of article counts by source.
    """
    plt.figure(figsize=(6,4))
    df["source"].value_counts().plot(kind="bar", color="orange", edgecolor="black")
    plt.title("Articles by Source")
    plt.xlabel("Source")
    plt.ylabel("Count")
    plt.show()
```
